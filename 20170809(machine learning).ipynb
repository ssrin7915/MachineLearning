머신러닝 - 선형회귀

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.datasets import load_boston
import pandas as pd
```

```python
df = sklearn.datasets.load_boston()
df
```

```python
x = df['data']
y = df['target']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
reg = linear_model.LinearRegression()
reg.fit(x_train,y_train),reg.intercept_,reg.coef_
```

```python
reg.predict(x_test[0,:])
```

```python
y_test[0]
```

```python
x_predict = reg.predict(x_test)
mean_squared_error(y_test,x_predict)
```

```python
r2_score(y_test,x_predict)
```

```python
pred_err = []
intercept = []
coef = []

for _ in range(10):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
    reg = linear_model.LinearRegression()
    reg.fit(x_train,y_train)
    x_predict = reg.predict(x_test)
    pred_err.append(mean_squared_error(y_test,x_predict))
    intercept.append(reg.intercept_)
    coef.append(reg.coef_)
```

```python
pred_err
```
