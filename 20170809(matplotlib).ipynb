```python
x1 = np.linspace(0.0, 5.0) 
x2 = np.linspace(0.0, 2.0) 
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) 
y2 = np.cos(2 * np.pi * x2) 

ax1 = plt.subplot(2, 1, 1) #2    
plt.plot(x1, y1, 'yo-') 
plt.title('A tale of 2 subplots') 
plt.ylabel('Damped oscillation') 
print(ax1) 

ax2 = plt.subplot(2, 1, 2) 
plt.plot(x2, y2, 'r.-') 
plt.xlabel('time (s)') 
plt.ylabel('Undamped') 
print(ax2) 
plt.show()
```

```python
men = [20, 35, 30, 35, 27]
women = [25, 32, 34, 20, 25]
x = np.arange(5)

xlabel = ['A', 'B', 'C' ,'D','E']

plt.bar(x, men,0.3,color='b' ,label='men')
plt.bar(x+0.3,women,0.3,color='r' ,label='women')
plt.xticks(x+0.15,xlabel)

plt.legend() #label
plt.title('Scores by group and gender')
plt.xlabel('Group')
plt.ylabel('Scores')
plt.show()
```
